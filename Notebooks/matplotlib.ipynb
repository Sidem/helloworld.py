<h1>Pyplot tutorial</h1>

```python
plt.plot([1,2,3,4,10])
plt.xlabel('The input that i did not specify')
plt.ylabel('The input that i spciefied as a python list')
plt.title('My fist plot')
plt.text(0.5,4,'This is positon (0.5,4)')
plt.grid(True)
plt.show()
# xlabel is used to write a description for the x axis
# ylabel to write description for the y-axis
# title is used to give title to the graph
# grid is used when you want to diaplay a grid on the graph
# text is used when you want to add text to the graph at the 
# specified location in the first tow parameters (in our case
# it is (0.5,4)).
# show is used to show the plt in the end
```

```python
#To plot multiple plots you can specify them in the same plot command
import numpy as np
t = np.arange(0.,4.,0.02)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

<h2>Working with multiple figures</h2>

```python
# Suppose you want to make a figure in which you want you want to plot
# the graph of each class. Now all the graphs belong to the same figure. 
# We specify our figure using figure(a) the value of a is just the 
# the id of the figure. Now to plot the subgraphs we use the command subplot
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0., 5., 0.1)
t2 = np.arange(0., 5., 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2),'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```

<p>Subplot has three parameters numrows,numcols,fignum where fignum ranges from 1 to numrows x numcols. The commas in the subplot are optional if numrows*numcol < 10 </p>
<p>So the 2 in the above example specifies the number of rows you want. In our case we chose 2 rows so we get the graph in 2 rows. Now the next 1 specifies the number of cols. If we made it 2 then both the plots would appear in the same row. The last number specifies the the position at which we want to display our subplot</p>

```python
plt.figure(1)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2),'k')

plt.subplot(224)
plt.plot(t2, np.cos(2*np.pi*t2),'r--')
plt.show()
```

<p>You can also write something like R<sub>i</sub>=15 in a grapg using text command. To know more about how to write them see the documentation. They come under the topic Tex equation expressions</p>

```python
# To change the linewidth of the lines of the plot use linewidth
plt.figure(1)
plt.subplot(221)
plt.plot([1,2,3,4],linewidth=3.0)

plt.subplot(222)
plt.plot([1,2,3,4])

plt.subplot(223)
plt.plot([1,2,3,4],lw=10)
plt.show()
```
