<p>Different placeholders:

```python
a = np.zeros((4,6))
b = np.ones((3,3), dtype=np.float64)
c = np.empty((3,3))
d = np.eye(4)
print (a,"\n")
print (b,"\n")
print (c,"\n")
print (d)
```

```python
# Now if you are given a array and you want another placeholder array of the same dimension
# you can use zeros_like,ones_like
x = np.arange(20).reshape(4,5)
a = np.zeros_like(x)
b = np.ones_like(x)
c = np.empty_like(x, dtype=complex)
print ("a=\n",a,"\n")
print ("b=\n",b,"\n")
print ("c=\n",c)
```

<h1>Indexing, Slicing and Iterating</h1>

```python
# Now the arrays defined in numpy are operated as lists 
# Therefore you can access the elements of array in the same way as a python list
a = np.arange(10)
print (a,"\n")
print ("a[8]=",a[8])
print ("\na[2:5]=",a[2:5])
```

```python
# you can also do slicing with certain jumps or breaks
print ("a[:6:2]=",a[:6:2])
# As it starets with the starting element then it has to go 
# till 6 where 6 is in open range and with a jump of 2
```

```python
# Now for multidimensional arrays
# These arrays can have one index per axis as in numpy dimensionis referred to as axis.
# Now the maximum number of axis is same as the rank of the array
# So the index for multidimensional array is given as a tuple
a = np.arange(20).reshape(4,5)
print ("a=\n",a)
print ("\na[2,3]=",a[2,3])
```

```python
# Now the slicing in multidimensioal arrays is same as single dimensioal array
# As we are now considering tuples of the form (a,b) here the slicin can be done as
# [:a,:b] or [3:a,4:b,3]
print (a[0:5,1])
```

```python
# If this is getting confusing remember this 
# a corresponds to rows and b corresponds to columns
# so 0:5 in the above example means 0 row to 5-1 row and 
# 1 means the 1 column
# Now another thing to note here is that the indexing starts from 0
# So when i say first column i mean second column as the first column is indexed as 0=
```

```python
# when lesser number of indexes are provided the rest are assumend to be complete slices
# or : in python language
print (a[1])
# here it wanted two indexes but we gave only one so it assumed the
# missing slice to be complete slice
```

```python
print (a[-1])
# If you are familiar with negative indexing then in numpy it is the exact same as in python
print (a[-2,-4])
```

<p>The expression within brackets in a[i] is treated as an i followed by as many instances of : as needed to represent the remaining axes. NumPy also allows you to write this using dots as a[i,...].</p>
<p>The dots (...) represent as many colons as needed to produce a complete indexing tuple. For example, if x is a rank 5 array (i.e., it has 5 axes), then
<uL>
<li>x[1,2,...] is equivalent to x[1,2,:,:,:]</li>
<li>x[...,3] to x[:,:,:,:,3]</li>
<li>x[4,...,5,:] to x[4,:,:,5,:]</li></ul></p>
